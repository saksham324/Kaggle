```python
import pandas as pd 
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/justmarkham/scikit-learn-videos/master/data/Advertising.csv', index_col=0)
```

```python
import seaborn as sns

%matplotlib inline
```

```python
sns.pairplot(data, x_vars=['TV', 'Radio', 'Newspaper'], y_vars=['Sales'], size=10, aspect=1, kind='reg')
```

```python
feature_cols = ['TV', 'Radio', 'Newspaper']
X = data[feature_cols]
y = data['Sales']
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
#Default split is 75% train and 25% test 
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression 
linreg = LinearRegression()
linreg.fit(X_train,y_train)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
zip(feature_cols, linreg.coef_)
```

```python
y_pred = linreg.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.mean_absolute_error(y_test, y_pred))
```

```python
import numpy as np 
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
feature_cols =['TV', 'Radio']
X = data[feature_cols]
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
linreg = LinearRegression()
linreg.fit(X_train,y_train)
y_pred = linreg.predict(X_test)
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
